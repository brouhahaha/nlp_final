```
import pandas as pd
import nltk
nltk.download('stopwords')
import numpy as np
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from pymorphy2 import MorphAnalyzer
from pymystem3 import Mystem

morph = MorphAnalyzer()
token = RegexpTokenizer('\w+')
stops = set(stopwords.words('russian'))

def normalize_pm(text):
    words = [morph.parse(word)[0].normal_form for word in tokenize(text) if word]
    return words

def tokenize(text):
    return token.tokenize(text)
```

Обучаем word2vec на корпусе отзывов

```
data = gensim.models.word2vec.LineSentence('/content/drive/My Drive/автобрея/corpus.txt')
```

```
model = gensim.models.Word2Vec(data, size=100, window=10, min_count=2, sg=0)
```

```
len(model.wv.vocab)
```

Считаем similarity слов из отзыва со словами из топиков

```
d_food_dups = {}
d_service_dups = {}

for file in ['3385.txt', '5037.txt', '6668.txt', '9036.txt', '24916.txt', '26284.txt', '26330.txt', '30808.txt', '37516.txt']:
  file_path = '/content/drive/My Drive/автобрея/' + file
  with open(file_path, 'r', encoding = 'utf-8') as f:
    text = f.read()
    sents = re.split('(?<=[.!?]) (?=[A-ЯЁA-Z])', text)

    for sent in sents:
      for word in sent.split():
        norm = ' '.join(normalize_pm(word))
        if norm not in stops:
          

          for f in food_topic:
            if norm in model.wv.vocab and f in model.wv.vocab:
              sim = model.similarity(norm, f)
              d_food_dups[(norm, sim)] = sent
        
          for s in service_topic:
            if norm in model.wv.vocab and s in model.wv.vocab:
              sim = model.similarity(norm, s)
              d_service_dups[(norm, sim)] = sent
```

```
for_food = {}
for k, value in d_food_dups.items():
  if k[0] in for_food:
    for_food[k[0]].append(k[1])
  else:
    for_food[k[0]] = [k[1]]

d_food = {}
for k, v in for_food.items():
  d_food[(k, max(v))] = d_food_dups[((k, max(v)))]
```

```
with open('/content/drive/My Drive/автобрея/food_candidates.txt', 'w', encoding = 'utf-8') as f:
  for k, v in d_food.items():
    f.write(str(k) + ' : '+ str(v) + '\n')
```
